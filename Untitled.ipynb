```python
import pandas as pd
from flaml import AutoML
import numpy as np
```

```python
data = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

# Разбиваем трейн на два датасета, сначала обучаем на том, где price_type = 1

```python
data_approved = data[data.price_type == 1]
```

```python
data_not_approved = data[data.price_type == 0]
```

```python
X_train = data_approved.drop(['per_square_meter_price'], axis=1)
y_train = data_approved.per_square_meter_price
```

```python
automl = AutoML()
```

```python
#используем кроссвалидацию с 4 фолдами
automl_settings = {
    "time_budget": 60 * 11,  # in seconds
    "metric": 'r2',
    "task": 'regression',
    "n_splits" : 4,
    "estimator_list" : ['lgbm', 'xgboost', 'rf', 'extra_tree'],
    "eval_method" : "cv",
    "seed" : 25,
    "n_jobs" : -1
}
```

```python
automl.fit(X_train=X_train, y_train=y_train, **automl_settings)
```

# (Промежуточный сабмит)

```python
y = automl.predict(test)
sub = pd.DataFrame({"id" : test.id, "per_square_meter_price" : y})
sub.to_csv('sub.csv', index=False)
```

```python
y
```

# (Продолжаем)
# Затем добавляем предсказанную цену во все датасеты

```python
test["predicted_price"] = automl.predict(test)
```

```python
y = automl.predict(data_not_approved.drop(['per_square_meter_price'], axis=1))
```

```python
data_not_approved["predicted_price"] = y
```

```python
data_approved["predicted_price"] = automl.predict(data_approved.drop(['per_square_meter_price'], axis=1))
```

# Отбираем только те значения, где данные о цене отличаются от предикта меньше чем на 10%

```python
data_approved = data_approved.append(data_not_approved.loc[abs(
    data_not_approved.per_square_meter_price - data_not_approved.predicted_price
    ) < data_not_approved.per_square_meter_price * 0.1])
```

```python
data_approved
```

# И снова обучаем модель на обновленных данных

```python
X_train = data_approved.drop(["per_square_meter_price"], axis=1)
y_train = data_approved.per_square_meter_price
```

```python
automl_final = AutoML()
```

```python
automl_settings = {
    "time_budget": 60 * 45,  # in seconds
    "metric": 'r2',
    "task": 'regression',
    "n_splits" : 4,
    "estimator_list" : ['lgbm', 'xgboost', 'rf', 'extra_tree'],
    "eval_method" : "cv",
    "seed" : 25,
    "n_jobs" : -1
}
```

```python
automl_final.fit(X_train=X_train, y_train=y_train, **automl_settings)
```

```python
y = automl_final.predict(test)
```

```python
y
```

```python
sub = pd.DataFrame({"id" : test.id, "per_square_meter_price" : y})
```

```python
sub.to_csv('sub.csv', index=False)
```
